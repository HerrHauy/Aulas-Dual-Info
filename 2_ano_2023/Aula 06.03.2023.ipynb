```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import curve_fit
```

```python
plt.plot([0,1,2,3],[0,0,0,0], 'o', color='blue')
plt.plot([4,5,6,7],[1,1,1,1], 'x', color='red')
plt.xlabel('Tamanho do tumor [cm]')
plt.ylabel('benigno/maligno')
plt.xlim(0,7)
plt.ylim(-0.5,1.5)
plt.show()
```

```python
xdata = [0,1,2,3,4,5,6,7]
ydata = [0,0,0,0,1,1,1,1]

def sigmoid(x, x0, k):
    y = 1 / (1 + np.exp(-k*(x-x0)))
    return y

popt, pcov = curve_fit(sigmoid, xdata, ydata)
print(popt)
```

```python
x_space = np.linspace(0,7,20)
val_sigma = []
popt, pcov = curve_fit(sigmoid, xdata, ydata)



for i in range(0,len(x_space)):
    val_sigma.append(sigmoid(x_space[i],popt[0],popt[1]))

plt.plot(xdata,ydata, 'o', color='blue')
plt.plot(x_space,val_sigma,'-', color='red')
plt.plot([popt[0],popt[0]],[-0.5,1.5],'--', color='green')
plt.xlabel('Tamanho do tumor [cm]')
plt.ylabel('benigno/maligno')
plt.xlim(0,7)
plt.ylim(-0.5,1.5)
plt.show()

print(f'Ponto de limite de decisão: {popt[0]}')
```

```python
xdata = [0,1,2,3,4,5,6,7,50]
ydata = [0,0,0,0,1,1,1,1,1]

x_space = np.linspace(0,50,100)
val_sigma = []
popt, pcov = curve_fit(sigmoid, xdata, ydata)



for i in range(0,len(x_space)):
    val_sigma.append(sigmoid(x_space[i],popt[0],popt[1]))

plt.plot(xdata,ydata, 'o', color='blue')
plt.plot(x_space,val_sigma,'-', color='red')
plt.plot([popt[0],popt[0]],[-0.5,1.5],'--', color='green')
plt.xlabel('Tamanho do tumor [cm]')
plt.ylabel('benigno/maligno')
plt.xlim(0,50)
plt.ylim(-0.5,1.5)
plt.show()

print(f'Ponto de limite de decisão: {popt[0]}')
```

```python
def previsao(x):
    if x<popt[0]:
        return f'Seu tumor provavelmente é benigno :)'
    elif x>popt[0]:
        return f'Seu tumor provavelmente é maligno :('
    else :
        f'Não sei o que fazer com isso.'
```
